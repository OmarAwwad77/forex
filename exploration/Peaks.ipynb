```python
import pandas as pd
from datetime import datetime
from dateutil import parser
from plotting import CandlePlot
import plotly.graph_objects as go
```

```python
start = parser.parse('2023-01-02T00:00:00Z')
end = parser.parse('2023-04-03T22:00:00Z')

df_h1 = pd.read_pickle(f"../data/candles/EUR_USD_H1.pkl")
df_m5 = pd.read_pickle(f"../data/candles/EUR_USD_M5.pkl")

df_h1 = df_h1[(df_h1.time >= start) & (df_h1.time <= end)].reset_index(drop=True)
df_m5 = df_m5[(df_m5.time >= start) & (df_m5.time <= end)].reset_index(drop=True)
```

```python
def show_plot(df: pd.DataFrame, trend_lines=[]):
    cp = CandlePlot(df, candles=True)
    for (x1, x2), y in trend_lines:
        x = [cp.df_plot.iloc[x1]['sTime'], cp.df_plot.iloc[x2]['sTime']]
        cp.fig.add_trace(go.Scatter(x=x, y=y))
    cp.show_plot(width=1100, height=500)
    
perc = 0.00455
price = 1.06779
show_plot(df_h1.iloc[:500], [[(26, 31), (price, price - price * perc )], [(10, 13), (price, price + price * 0.0007 )]])
```
